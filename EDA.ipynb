### *EDA (Exploratory Data Analysis)*

First we start importig all the libraries that we need to our notebook.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%config IPCompleter.greedy = True
```

We read the Dataset and show a few lines of it.

```python
df = pd.read_csv("creditcard.csv")
df.head()
```

We show the shape of the dataframe (number of rows and columns) and some descriptive info about the data with info command 

```python
df.shape
```

```python
df.info()
```

As we can see, we do not need to impute missing values because there are not null values in the dataframe.

In order to continue with the explore the data, we display some information with the describe function.

```python
df.describe()
```

Let's find out the proportion in Class column.

```python
counts = df.Class.value_counts()
counts
```

```python
print("Percentage of Negative Fraud: {0:.4f}%".format(counts[0]/df.shape[0]))
print("Percentage of Positive Fraud: {0:.4f}%".format(counts[1]/df.shape[0]))
```

```python
sns.countplot(x = df.Class.values)
```

The dataset is unbalanced.

Now we can ask ourselves if the 'Amount' of fraudulent transaction are higher than non fraudulent. In order to do that, first we explore the 'Amount' column distribution.

```python
df['Amount'].hist(figsize=(10,10), bins=100)
```

```python
df[(df.Class == 0)].Amount.plot.hist(title = 'Histogram of valid transactions')
```

```python
df[df.Class == 1].Amount.plot.hist(title = 'Histogram of fraudulent transactions')
```

```python
plt.rcParams['figure.figsize'] = (12,5)
sns.boxplot(y='Amount', x='Class', data=df)
```

We can see that both distributions look very similar. So we cannot say that fraudulents amount are higher than not fraudulent.

Now we can investigate how the both transactions (fraudulent and the not ones) are related over time

```python
plt.figure()
plt.axes()
sns.scatterplot(data=df, x='Time', y='Amount',style='Class',hue='Class',size='Class',sizes=(75, 25))
plt.show()
```

Finally we show the correlation between all the varaibles.

```python
corr = df.corr('pearson')
plt.figure(figsize=(20,20))
sns.heatmap(corr, cbar = True,  square = True, annot=True, fmt= '.2f',annot_kws={'size': 10},
           xticklabels= df.columns, 
           yticklabels= df.columns,
           cmap= 'coolwarm')
plt.show()
```

Variables V1 to V28 are totally uncorrelated. That is not the case for Time and Amount columns.
